```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
we_rate_dogs = pd.read_csv('Twitter_archive_master.csv')
```

Insight 1: The most common dog breed is the golden_retriever. This is of course according to the first prediction of the neural network as depicted in the bar chart below.

```python
we_rate_dogs['first_prediction'].value_counts()[:10].plot(kind = 'bar', figsize = (20,5), rot = 0);
plt.xlabel('Dog Names');
plt.ylabel('Frequency');
plt.title('Most common dog breeds')
plt.legend();
```

### Insight 2:
Pupper is the highest occuring dog stage in the sample of tweets. could this be said for all dogs? well!

```python
we_rate_dogs['dog_stage'].value_counts().plot(kind = 'bar', figsize = (10,5), rot = 0);
plt.xlabel('Dog Stage');
plt.ylabel('Frequency');
plt.title('Most common dog Stage')
plt.legend();
```

### Insight 3: 
With the exception of 'a' and 'the, Charlie, Lucy, Cooper, Oliver, Penny, Tucker, Sadie, Winston, Toby, Daisy, Lola are the 10 most common dog names

```python
we_rate_dogs['name'].value_counts()[1:15].plot(kind = 'bar', figsize =(20,10), rot = 0);
plt.xlabel('Dog Names');
plt.ylabel('Frequency');
plt.title('Most common dog Names')
plt.legend();
```
